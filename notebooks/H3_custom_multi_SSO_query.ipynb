```python
# EPHESSOS - H3_custom_multi_SSO_query.ipynb
# A. S. Borlaff - NASA/STA - April 29, 2026
# 
# The objective of this notebook is to demonstrate the capability to:
# 1 - Load a MPC format table with oscullating orbital elements from multiple SSOs (real or simulated) 
# 2 - Send the query to JPL/Horizons and retrieve the ephemerides. 
#
# The precision of the retrieved solutions was done in H1_Eros_from_JPL_Horizons. 
# The retrieved ephemerides are good down to 0.1 - 0.3 arcsec. 
```

```python
import ephessos as ep
from astropy.time import Time
import astropy.units as u

now = Time(60000, format="mjd") # Time.now()
mjd = now.mjd # Time((mjd_end - mjd_start)/2., format="mjd")
mjd_start = mjd - 100
mjd_end = mjd + 100

mpc_table = "NEA-test.txt"
ephessos_df = ep.core.ephessos(mpc_table=mpc_table, mjd_start=mjd_start, mjd_end=mjd_end, step_size="1d", verbose=False)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,8))

for i in range(len(ephessos_df)):
    ax.plot(ephessos_df[i]["RA_deg_ICRF"], ephessos_df[i]["DEC_deg_ICRF"], label=ephessos_df[i]["Designation"][0])

ax.set_ylabel("Declination (degrees)")
ax.set_xlabel("Right ascension (degrees)")

plt.legend() 
plt.savefig("H3_custom_multi_SSO_query.png", dpi=300)
plt.show()
```
